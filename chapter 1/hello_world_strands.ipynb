### Configuring AWS Credentials in Colab

To resolve the `NoCredentialsError`, you need to provide your AWS credentials to the Colab environment. The recommended way is to use Colab's secret manager for sensitive information.

1.  **Open the Secrets tab:** Click the "🔑" icon in the left-hand panel of your Colab notebook.
2.  **Add new secrets:** Add the following three secrets, pasting your actual AWS values:
    *   `AWS_ACCESS_KEY_ID`
    *   `AWS_SECRET_ACCESS_KEY`
    *   `AWS_REGION` (e.g., `us-east-1`)
3.  **Ensure "Notebook access" is enabled** for these secrets.

Once configured, you can load these secrets into your environment like this:

```python
from google.colab import userdata
import os

# Retrieve secrets from Colab's secret manager
os.environ['AWS_ACCESS_KEY_ID'] = userdata.get('AWS_ACCESS_KEY_ID')
os.environ['AWS_SECRET_ACCESS_KEY'] = userdata.get('AWS_SECRET_ACCESS_KEY')
os.environ['AWS_REGION'] = userdata.get('AWS_REGION')

print("AWS credentials loaded successfully from Colab secrets.")
```

After running the cell above, the `strands` agent should be able to find your AWS credentials and connect to Amazon Bedrock. Now you can try running the agent again.
